## Task 2a: Indexing by Subsetting and Slicing

In the practice notebook perform the following:

1. Create (or re-use) 3 arrays, each containing three dimensions.
2. Slice each of these arrays so that:
    + One element / number is returned.
    + One dimension is returned.
    + A subset of a dimension is returned.
3. What is the difference between `[x:]` and `[x, ...]`? (hint, try each on high-dimension arrays).
    
*Exactly what you choose to return is not imporant at this point, the goal of this task is to train you so that if you are given an n-dimension NumPy array, you can write an index or slice that returns a subset of desired positions.*

```python
#creating three 3D arrays
arr1 = np.random.random((3, 3, 3))
arr2 = np.array([[[1, 2, 3], [5, 6, 7], [7, 8, 9], [10, 11, 12]]])
arr3 = np.ones((3, 3, 3))
print(arr1)
print(arr2)
print(arr3)

#returning one element
one_ele1 = arr1[1, 2, 1]
one_ele2 =arr2[0, 1, 1]
one_ele3 =arr3[2, 1, 2]
print(one_ele1)
print(one_ele2)
print(one_ele3)

#returning one dimension
one_dim1 = arr1[:,1, 1]
one_dim2 = arr2[-1, :, -1]
one_dim3 = arr3[1, :, 1] 
print(one_dim1)
print(one_dim2)
print(one_dim3)

#returning subset of a dimension
sub1 = arr1[1:2,1, 1]
sub2 = arr2[-1, 0:1, 0]
sub3 = arr3[1:2, 0:, 1] 
print(sub1)
print(sub2)
print(sub3)

print(arr1[1:])
print(arr1[1, ...])

print(arr2[:2:])
print(arr2[..., 2])
```

## Task 3a: Boolean Indexing

In the practice notebook perform the following:

+ Experiment with the following boolean conditionals to generate boolean arrays for indexing:
  + Greater than
  + Less than
  + Equals
  + Combine two or more of the above with:
      + or `|`
      + and `&`

You can create arrays or use existing ones

```python
array1 = np.random.random((5, 5))
array2 = np.random.random((4, 3)) 
array3 = np.random.random((6, 3))

print(array1)
print(array2)
print(array3)

#all numbers greater than 0.3 in array 1
print(array1 > 0.3)

#all numbers greater than 0.5 in array 2
print(array2 > 0.5)

#all numbers greater than 0.8 in array 3
print(array3 > 0.8)


#all numbers less than 0.4 in array 1
print(array1 < 0.4)

#all numbers less than 0.5 in array 2
print(array2 < 0.5)

#all numbers less than 0.1 in array 3
print(array3 < 0.1)


#all numbers equal to 0.5 in array 1
print(array1 == 0.5)

#all numbers equal to 0.2 in array 2
print(array2 == 0.2)

#all numbers equal to 0.44 in array 3
print(array3 == 0.44)

#numbers less than 0.5 and more than 0.2 in array1
print((array1 > 0.2) & (array1 < 0.5))

array4 = np.random.random((3, 3))
print(array4)

#numbers more than 0.5 and less than 0.9 in array 4
print((array4 > 0.5) & (array4 < 0.9))

#numbers more than 0.3 or less than 0.8 in array 4
print((array4 > 0.5) | (array4 < 0.9))
```
